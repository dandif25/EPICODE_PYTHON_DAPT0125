```python
import pandas as pd

df_covid = pd.read_csv("C:/Users/UTENTE/Desktop/CORSO/Esercizi_PYTHON/dataset/datasets/beginner_datasets/owid-covid-data.csv")

df_covid.info()
df_covid.head()
```

```python
# Rimuovi righe con continent NaN (es. voci globali come "World", "International", ecc.)
df_covid = df_covid[df_covid["continent"].notna()]

# Ordina per data e prendo l'ultima riga per ogni paese (dove total_cases è aggiornato)
df_latest = df_covid.sort_values("date").groupby("location", as_index=False).last()

# Raggruppo per continente e somma i total_cases
continent_cases = df_latest.groupby("continent")["total_cases"].sum().reset_index()

# Calcola il totale mondiale
total_global_cases = continent_cases["total_cases"].sum()

# Aggiungo colonna percentuale
continent_cases["percentuale_sul_totale"] = (continent_cases["total_cases"] / total_global_cases) * 100

# Arrotondo la percentuale a due cifre decimali
continent_cases["percentuale_sul_totale"] = continent_cases["percentuale_sul_totale"].round(2)

continent_cases
```

```python
# Converto la colonna "date" in datetime
df_covid["date"] = pd.to_datetime(df_covid["date"])

# Seleziono dei dati per l'Italia nel 2022
italy_2022 = df_covid[
    (df_covid["location"] == "Italy") &
    (df_covid["date"].dt.year == 2022)
]

# Filtro solo le righe dove ci sono nuovi casi registrati (non nulli e maggiori di zero)
italy_2022_filtro = italy_2022[italy_2022["new_cases"].notna() & (italy_2022["new_cases"] > 0)]

italy_2022_filtro
```

```python
import matplotlib.pyplot as plt

# A. GRAFICO: Evoluzione dei casi totali nel 2022
plt.figure(figsize=(12, 5))
plt.plot(italy_2022_filtro["date"], italy_2022_filtro["total_cases"], marker='o', linestyle='-')
plt.title("Evoluzione dei casi totali COVID-19 in Italia - 2022")
plt.xlabel("Data")
plt.ylabel("Casi totali")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# B. GRAFICO: Nuovi casi Covid-19 2022
plt.figure(figsize=(12, 5))
plt.bar(italy_2022_filtro["date"], italy_2022_filtro["new_cases"], width=3, color="salmon")
plt.title("Nuovi casi COVID-19 in Italia - 2022")
plt.xlabel("Data")
plt.ylabel("Nuovi casi")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Converto la colonna "date" in datetime
df_covid["date"] = pd.to_datetime(df_covid["date"])

# Seleziono i tre paesi
paesi = ["Italy", "Germany", "France"]
inizio_date = "2022-05-01"
fine_date = "2023-04-30"

df_covid = df_covid[
    (df_covid["location"].isin(paesi)) &
    (df_covid["date"] >= inizio_date) &
    (df_covid["date"] <= fine_date)
]

# Rimuovo le righe con valori NaN nella colonna 'icu_patients'
df_covid = df_covid[df_covid["icu_patients"].notna()]

df_covid
```

```python
import seaborn as sns

# A. Boxplot della colonna 'icu_patients'
plt.figure(figsize=(10, 6))
sns.boxplot(x="location", y="icu_patients", data=df_covid)
plt.title("Pazienti in terapia intensiva (ICU) da Maggio 2022 a Aprile 2023")
plt.xlabel("Paese")
plt.ylabel("Pazienti in ICU")
plt.grid(True)
plt.tight_layout()
plt.show()
```
