```python
import pandas as pd

df_diabetes = pd.read_csv("C:/Users/UTENTE/Desktop/CORSO/Esercizi_PYTHON/dataset/datasets/beginner_datasets/diabetes.csv")

df_diabetes.info()
df_diabetes.head(5)
```

```python
df_diabetes.describe()
```

```python
df_diabetes["Age (years)"] #seleziono la colonna dal data frame che mi interessa
filtro_under20 = df_diabetes["Age (years)"] < 20
filtro20_30 = (df_diabetes ["Age (years)"] >= 20) & (df_diabetes["Age (years)"] < 30)
filtro30_40 = (df_diabetes["Age (years)"] >= 30) & (df_diabetes ["Age (years)"] < 40)
filtro40_50 = (df_diabetes["Age (years)"] >= 40) & (df_diabetes ["Age (years)"] < 50)
filtro_over50 = df_diabetes ["Age (years)"] >= 50 #creo un filtro in cui da questa colonna prende sono i maggiori di 50


df_diabetes[filtro_over50] #applico al dataframe il filtro
```

```python
# Media pressione sanguigna per diverse fasce di età
print("Minore di 20:", df_diabetes[filtro_under20]["Diastolic blood pressure (mm Hg)"].mean())
print("Da 20 a 30:  ", df_diabetes[filtro20_30]["Diastolic blood pressure (mm Hg)"].mean())
print("Da 30 a 40:  ", df_diabetes[filtro30_40]["Diastolic blood pressure (mm Hg)"].mean())
print("Da 40 a 50:  ", df_diabetes[filtro40_50]["Diastolic blood pressure (mm Hg)"].mean())
print("Maggiore di 50:", df_diabetes[filtro_over50]["Diastolic blood pressure (mm Hg)"].mean())
```

```python
anni = df_diabetes["Age (years)"].unique() # Salvo i valori unici di age in una variabile

anni.sort() # Ordino una variabile che si aggiorna

for anno in anni:
    media_pressione = df_diabetes[df_diabetes["Age (years)"] == anno]["Diastolic blood pressure (mm Hg)"].mean()
    print("Anno",anno,"- media pressione:", round(media_pressione,2))
```

```python
df_pokemon = pd.read_csv("C:/Users/UTENTE/Desktop/CORSO/Esercizi_PYTHON/dataset/datasets/beginner_datasets/pokemon.csv")

df_pokemon.info()
```

```python
# È verosimile che la prima colonna dovrebbe essere un indice? 
# No?!
```

```python
# Confrontiamolo con l'indice messo automaticamente da Pandas: combaciano?
df_pokemon.loc[:,"#"] # : seleziona tutte le righe "#" è la colonna
```

```python
# Settare la prima colonna "#" come indice
df_pokemon.set_index("#")
```

```python
# Quali sono i Pokémon leggendari?
leggendari = df_pokemon["Legendary"] == True

pokemon_leggendari = df_pokemon[leggendari][["#","Name"]]

pokemon_leggendari.set_index("#")
```

```python
# E quali sono i leggendari di tipo 1 Grass?
filtro_tipo = (df_pokemon["Type 1"] == "Grass") & leggendari

pokemon_grass = df_pokemon[filtro_tipo][["#","Name"]]

pokemon_grass.set_index("#")
```

```python
# E leggendari di tipo 1 Ice o Fire?
filtro_ice = df_pokemon["Type 1"] == "Ice"
filtro_fire = df_pokemon["Type 1"] == "Fire"

filtro_ice_fire = filtro_ice | filtro_fire

pokemon_ice_o_fire = df_pokemon[filtro_ice_fire & leggendari]

pokemon_ice_o_fire[["#","Name","Type 1"]]
```

```python
# Ordiniamo il dataset per la colonna Name;
df_pokemon = df_pokemon.sort_values(by = "Name")
df_pokemon
```

```python
# Trasformiamo Name nell'indice;
df_pokemon.set_index("Name")
```

```python
# Quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?
filtro_generazione = df_pokemon["Generation"] == 1
filtro_attacco = df_pokemon["Attack"] > 50
filtro_hp = df_pokemon["HP"] < 60

filtro_completo = filtro_generazione & filtro_attacco & filtro_hp

df_pokemon[filtro_completo]
```
